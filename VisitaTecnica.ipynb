**1. ¿En qué consiste el trabajo de un Ing Sr OYM Core e Ingenieria de Red?**

Un ingeniero Senior OYM Core e Ingeniería de Red en el ámbito de las telecomunicaciones cumple un rol clave en el diseño, implementación, operación y mantenimiento de la infraestructura de red, especialmente en el núcleo (core) de las redes de telecomunicaciones. 

Su trabajo consiste en: 
*1. Diseño e implementación* : Diseña arquitecturas de red robustas y escalables (LAN, WAN, WLAN, redes móviles), también define la topología de red e implementa soluciones de conectividad para garantizar eficiencia y disponibilidad. 

*2. Gestión del Core de Red* : Administra y optimiza los elementos del núcleo de la red (Core Network) y asegura la interoperabilidad entre diferentes tecnologías y servicios.

*3. Operación y mantenimiento* : Supervisa el rendimiento de la red en tiempo real, diagnostica y resuelve incidencias complejas que han sido escaladas por otros niveles de soporte. 

*4. Seguridad* : Implementas medidas de seguridad para proteger la red contra accesos no autorizados, ataques cibernéticos y fallos. 

*5. Documentación y cumplimiento* : Mantiene documentación técnica actualizada sobre configuraciones, cambios y procedimientos.
