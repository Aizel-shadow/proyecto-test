**1. En qu consiste el trabajo de un Ing Sr OYM Core e Ingenieria de Red?**

Un ingeniero Senior OYM Core e Ingeniera de Red en el mbito de las telecomunicaciones cumple un rol clave en el diseo, implementacin, operacin y mantenimiento de la infraestructura de red, especialmente en el ncleo (core) de las redes de telecomunicaciones. 

Su trabajo consiste en: 
*1. Diseo e implementacin* : Disea arquitecturas de red robustas y escalables (LAN, WAN, WLAN, redes mviles), tambin define la topologa de red e implementa soluciones de conectividad para garantizar eficiencia y disponibilidad. 

*2. Gestin del Core de Red* : Administra y optimiza los elementos del ncleo de la red (Core Network) y asegura la interoperabilidad entre diferentes tecnologas y servicios.

*3. Operacin y mantenimiento* : Supervisa el rendimiento de la red en tiempo real, diagnostica y resuelve incidencias complejas que han sido escaladas por otros niveles de soporte. 

*4. Seguridad* : Implementas medidas de seguridad para proteger la red contra accesos no autorizados, ataques cibernticos y fallos. 

*5. Documentacin y cumplimiento* : Mantiene documentacin tcnica actualizada sobre configuraciones, cambios y procedimientos.
