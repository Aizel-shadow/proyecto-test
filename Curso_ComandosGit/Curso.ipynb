## COMANDOS BÁSICOS DE GIT  

| Comando                   | Descripción                                                                    |
| ------------------------- | ------------------------------------------------------------------------------ |
| `git init`                | Inicializa un nuevo repositorio local de Git.                                  |
| `git clone <url>`         | Clona un repositorio remoto (por ejemplo, de GitHub) a tu máquina local.       |
| `git status`              | Muestra el estado actual del repositorio (archivos modificados, staged, etc.). |
| `git add <archivo>`       | Agrega archivos al área de staging (preparación para commit).                  |
| `git add .`               | Agrega todos los archivos modificados al área de staging.                      |
| `git commit -m "mensaje"` | Guarda los cambios en el historial del repositorio con un mensaje.             |
| `git log`                 | Muestra el historial de commits.                                               |
| `git diff`                | Muestra los cambios entre versiones de archivos antes de hacer commit.         |

## TRABAJO CON RAMAS (BRANCHES)

| Comando                    | Descripción                                     |
| -------------------------- | ----------------------------------------------- |
| `git branch`               | Lista las ramas existentes.                     |
| `git branch <nombre>`      | Crea una nueva rama.                            |
| `git checkout <rama>`      | Cambia a otra rama.                             |
| `git checkout -b <nombre>` | Crea y cambia a una nueva rama al mismo tiempo. |
| `git merge <rama>`         | Fusiona otra rama con la rama actual.           |
| `git branch -d <rama>`     | Elimina una rama local.                         |

## TRABAJO CON REPOS REMOTOS (GitHub)

| Comando                       | Descripción                                                                 |
| ----------------------------- | --------------------------------------------------------------------------- |
| `git remote -v`               | Muestra los remotos configurados.                                           |
| `git remote add origin <url>` | Conecta tu repositorio local con un repositorio remoto (usualmente GitHub). |
| `git fetch`                   | Descarga cambios del repositorio remoto sin aplicarlos.                     |
| `git pull`                    | Descarga y fusiona cambios del repositorio remoto.                          |
| `git push origin <rama>`      | Envía tus commits a GitHub en una rama específica.                          |
| `git push -u origin <rama>`   | Igual que arriba, pero además configura esa rama como predeterminada.       |

## SUBMODULES, STASH, ETC 

| Comando                          | Descripción                                     |
| -------------------------------- | ----------------------------------------------- |
| `git stash`                      | Guarda temporalmente cambios no confirmados.    |
| `git stash apply`                | Recupera los cambios guardados con `stash`.     |
| `git submodule add <url> <ruta>` | Agrega un submódulo a tu proyecto.              |
| `git submodule update --init`    | Inicializa y actualiza submódulos.              |
| `git reset --hard`               | Revierte todos los cambios locales (¡cuidado!). |
| `git clean -fd`                  | Elimina archivos y carpetas sin seguimiento.    |

## REPOSITORIO FORK, ORIGIN, UPSTREAM

| Comando                         | Descripción                                               |
| ------------------------------- | --------------------------------------------------------- |
| `git remote add upstream <url>` | Agrega el repo original (usualmente el que hiciste fork). |
| `git fetch upstream`            | Descarga cambios del repo original (upstream).            |
| `git merge upstream/main`       | Fusiona cambios del upstream con tu rama local.           |
| `git pull upstream main`        | Lo anterior en un solo paso.                              |
